```python
import pandas as pd
import numpy as np 
import mplsoccer as mpl 
from mplsoccer import Pitch
import matplotlib.pyplot as plt
from statsbombpy import sb
```

get last of all match ID in the euros 

```python
matches = pd.DataFrame()
matches = sb.matches(55,282)
match_id_list = matches['match_id'].to_list()
match_id_list
```

Using the matchID list, we aim to create a list of players by iterating through lineups of each match and adding the unique values 

ISSUE: the team_id variable is unavailable here 

```python
master_lineup = pd.DataFrame()
for i in range(0,50):
    lineup = sb.lineups(match_id_list[i])
    for team in lineup.values():
        team_df = pd.DataFrame(team)
        master_lineup = pd.concat([master_lineup,team_df],ignore_index=True)
master_lineup = master_lineup.drop_duplicates(subset='player_id')
master_lineup
```

```python
#getting rid of the nickname and name mismatch 
def replace_name_with_nickname(row):
    if pd.notna(row['player_nickname']):
        return row['player_nickname']
    return row['player_name']
master_lineup['player_name'] = master_lineup.apply(replace_name_with_nickname, axis=1)

# Drop the player_nickname column
master_lineup = master_lineup.drop(columns=['player_nickname'])
master_lineup
```

Current objective : to obtain an aggregate of all events of a specific type player wise

 gather all event data across the tournament pertaining to a specific player in a different dataframe 


```python
events = sb.events(match_id= 3930159)
pid = 3500
player_event_df = events[events['player_id']==pid]
player_event_df
```

```python

# Player ID to filter events
pid = 3500

# Initialize an empty list to store dataframes
player_events_list = []

# Iterate over each match ID
for match_id in match_id_list:
    # Load events for the current match
    events = sb.events(match_id=match_id)
    
    # Filter events for the specified player
    player_event_df = events[events['player_id'] == pid]
    
    # Append the filtered dataframe to the list
    player_events_list.append(player_event_df)

# Concatenate all dataframes in the list into a single dataframe
all_player_events_df = pd.concat(player_events_list, ignore_index=True)

# Display the combined dataframe
all_player_events_df
```
