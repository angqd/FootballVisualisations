```python
import pandas as pd
import numpy as np 
import mplsoccer as mpl 
from mplsoccer import Pitch
import matplotlib.pyplot as plt
from statsbombpy import sb
```

The goal of this notebook will be to find the best progressive passers in the tournament on per 90 basis 
and then we can figure what visualisations we want to generate from that

```python
matches = pd.DataFrame()
matches = sb.matches(55,282)
match_id_list = matches['match_id'].to_list
match_id_list
```

what are the goals , first figure out a player, then extract all pass events with that player id per match and add that cumilative information to a dataframe that stores the overall data 

```python
master_lineup = pd.DataFrame()
for i in range(0,50):
    lineup = sb.lineups(match_id_list[i])
    for team in lineup.values():
        team_df = pd.DataFrame(team)
        master_lineup = pd.concat([master_lineup,team_df],ignore_index=True)
master_lineup = master_lineup.drop_duplicates(subset='player_id')
master_lineup.head()
```
